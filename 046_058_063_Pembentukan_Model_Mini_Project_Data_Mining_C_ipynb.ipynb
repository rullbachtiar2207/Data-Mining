### Pra-pemrosesan Data (Data Preprocessing)

```python
# Data Preprocessing

# Function to extract skills from profiles and job descriptions
def extract_skills(text):
    if not isinstance(text, str):
        return []

    # Define common tech skills (expand as needed)
    common_skills = [
        'python', 'java', 'javascript', 'sql', 'r', 'c\+\+', 'c#', 'ruby', 'php', 'swift',
        'tensorflow', 'pytorch', 'keras', 'scikit-learn', 'pandas', 'numpy', 'matplotlib',
        'ai', 'machine learning', 'deep learning', 'nlp', 'natural language processing',
        'data science', 'big data', 'data analytics', 'data visualization', 'data mining',
        'database', 'cloud', 'aws', 'azure', 'gcp', 'google cloud', 'docker', 'kubernetes',
        'devops', 'ci/cd', 'jenkins', 'git', 'github', 'gitlab', 'jira', 'agile', 'scrum',
        'frontend', 'backend', 'full stack', 'web development', 'mobile development',
        'react', 'angular', 'vue', 'node.js', 'express.js', 'django', 'flask', 'spring',
        'rest api', 'graphql', 'microservices', 'mongodb', 'mysql', 'postgresql', 'oracle',
        'nosql', 'hadoop', 'spark', 'kafka', 'tableau', 'power bi', 'excel', 'sas', 'spss',
        'statistics', 'linear algebra', 'calculus', 'probability', 'regression', 'classification',
        'neural networks', 'cnn', 'rnn', 'lstm', 'gan', 'computer vision', 'image processing',
        'security', 'cryptography', 'blockchain', 'ethereum', 'solidity', 'smart contracts',
        'unity', 'unreal engine', 'game development', 'ar', 'vr', 'augmented reality', 'virtual reality',
        'project management', 'leadership', 'communication', 'problem solving', 'critical thinking',
        'html', 'css', 'sass', 'less', 'bootstrap', 'tailwind', 'jquery', 'dom', 'xml', 'json',
        'rest', 'soap', 'api', 'testing', 'junit', 'selenium', 'cypress', 'marketing', 'seo', 'sem',
        'product management', 'ui/ux', 'figma', 'adobe xd', 'photoshop', 'illustrator', 'sketch',
        'data structures', 'algorithms', 'networking', 'tcp/ip', 'dns', 'linux', 'unix',
        'shell scripting', 'bash', 'powershell', 'windows', 'macos', 'android', 'ios', 'swift', 'kotlin'
    ]

    # Convert text to lowercase and search for skills
    text = text.lower()
    found_skills = []

    for skill in common_skills:
        # Use word boundaries to match complete words
        pattern = r'\b' + re.escape(skill) + r'\b'
        if re.search(pattern, text):
            found_skills.append(skill)

    return found_skills

# Function to clean and preprocess text
def preprocess_text(text):
    if not isinstance(text, str):
        return ""

    # Convert to lowercase and remove special characters
    text = text.lower()
    text = re.sub(r'[^\w\s]', ' ', text)

    # Tokenize and remove stopwords
    stop_words = set(stopwords.words('english'))
    tokens = word_tokenize(text)
    filtered_tokens = [w for w in tokens if w not in stop_words]

    return ' '.join(filtered_tokens)
```

### Fungsi untuk Membuat Fitur Keahlian (TF-IDF)

```python
# Create skill vocabulary and TF-IDF features
def create_skill_features(jobs_data, max_features=100):
    """Create TF-IDF features from job skills"""
    # Combine all skills into text format
    skill_texts = []
    for skills_list in jobs_data['skills']:
        if isinstance(skills_list, list) and len(skills_list) > 0:
            skill_texts.append(' '.join(skills_list))
        else:
            skill_texts.append('')

    # Create TF-IDF vectorizer
    vectorizer = TfidfVectorizer(max_features=max_features, ngram_range=(1,2))
    tfidf_matrix = vectorizer.fit_transform(skill_texts)

    return tfidf_matrix, vectorizer
```

### Fungsi untuk Membuat Graf Pekerjaan

```python
# Create graph from job similarity
def create_job_graph(similarity_matrix, threshold=0.1):
    """Create graph from similarity matrix"""
    n_jobs = similarity_matrix.shape[0]
    edges = []
    edge_weights = []

    # Create edges based on similarity threshold
    for i in range(n_jobs):
        for j in range(i+1, n_jobs):
            if similarity_matrix[i, j] > threshold:
                edges.append([i, j])
                edges.append([j, i])  # Undirected graph
                edge_weights.extend([similarity_matrix[i, j], similarity_matrix[i, j]])

    if len(edges) == 0:
        print(f"Warning: No edges found with threshold {threshold}. Creating minimum spanning tree...")
        # If no edges, create edges for each node to its most similar neighbor
        for i in range(n_jobs):
            similarities = similarity_matrix[i].copy()
            similarities[i] = -1  # Exclude self
            most_similar = np.argmax(similarities)
            if similarities[most_similar] > 0:
                edges.append([i, most_similar])
                edges.append([most_similar, i])
                edge_weights.extend([similarities[most_similar], similarities[most_similar]])

    return np.array(edges).T, np.array(edge_weights)
```

### Definisi Arsitektur Model GNN dan GAT

```python
# GNN Model for node embeddings
class JobGNN(nn.Module):
    def __init__(self, input_dim, hidden_dim=64, output_dim=32):
        super(JobGNN, self).__init__()
        self.conv1 = GCNConv(input_dim, hidden_dim)
        self.conv2 = GCNConv(hidden_dim, output_dim)
        self.dropout = nn.Dropout(0.2)

    def forward(self, x, edge_index, edge_weight=None):
        x = F.relu(self.conv1(x, edge_index, edge_weight))
        x = self.dropout(x)
        x = self.conv2(x, edge_index, edge_weight)
        return x

class JobGAT(nn.Module):
    def __init__(self, input_dim, hidden_dim=64, output_dim=32, heads=4):
        super(JobGAT, self).__init__()
        self.conv1 = GATConv(input_dim, hidden_dim // heads, heads=heads, dropout=0.2)
        self.conv2 = GATConv(hidden_dim, output_dim, heads=1, dropout=0.2)
        self.dropout = nn.Dropout(0.2)

    def forward(self, x, edge_index, edge_weight=None):
        x = F.relu(self.conv1(x, edge_index))
        x = self.dropout(x)
        x = self.conv2(x, edge_index)
        return x
```

### Fungsi untuk Mengevaluasi Kualitas Clustering

```python
# Clustering evaluation function
def evaluate_clustering(features, labels, data_name=""):
    """Evaluate clustering quality"""
    try:
        n_clusters = len(np.unique(labels))
        silhouette = silhouette_score(features, labels)
        calinski = calinski_harabasz_score(features, labels)
        davies_bouldin = davies_bouldin_score(features, labels)

        return {
            'n_clusters': n_clusters,
            'silhouette_score': silhouette,
            'calinski_harabasz_score': calinski,
            'davies_bouldin_score': davies_bouldin
        }
    except Exception as e:
        print(f"Error in evaluation for {data_name}: {str(e)}")
        return {
            'n_clusters': len(np.unique(labels)),
            'silhouette_score': -1,
            'calinski_harabasz_score': -1,
            'davies_bouldin_score': -1,
            'error': str(e)
        }
```

### Fungsi Utama untuk Menjalankan Eksperimen Clustering

```python
def run_clustering_experiment(features, feature_name, model_type="GNN", n_clusters=5):
    """Run clustering experiment with given features and model type"""
    print(f"\n--- {feature_name} + {model_type} Experiment ---")

    try:
        # Convert features to dense if sparse
        if hasattr(features, 'toarray'):
            features_dense = features.toarray()
        else:
            features_dense = features

        print(f"Feature shape: {features_dense.shape}")

        # Calculate similarity matrix
        similarity_matrix = cosine_similarity(features_dense)
        print(f"Similarity matrix shape: {similarity_matrix.shape}")
        print(f"Similarity range: [{similarity_matrix.min():.4f}, {similarity_matrix.max():.4f}]")

        # Create graph
        edge_index, edge_weights = create_job_graph(similarity_matrix, threshold=0.05)
        print(f"Graph edges: {edge_index.shape[1]}")

        if edge_index.shape[1] == 0:
            print("Warning: No edges in graph, using identity clustering")
            cluster_labels = np.arange(len(features_dense)) % n_clusters
            evaluation = evaluate_clustering(features_dense, cluster_labels, f"{feature_name}+{model_type}")
            return {
                'method': f"{feature_name} + {model_type}",
                'cluster_labels': cluster_labels,
                'evaluation': evaluation,
                'features_shape': features_dense.shape,
                'n_edges': 0,
                'error': 'No edges in graph'
            }

        # Create PyTorch geometric data
        x = torch.FloatTensor(features_dense)
        edge_index_tensor = torch.LongTensor(edge_index)
        edge_weight_tensor = torch.FloatTensor(edge_weights)

        # *** PILIH MODEL BERDASARKAN model_type ***
        if model_type == "GNN":
            model = JobGNN(input_dim=features_dense.shape[1])
        elif model_type == "GAT":
            model = JobGAT(input_dim=features_dense.shape[1])
        else:
            raise ValueError(f"Unknown model type: {model_type}")

        optimizer = torch.optim.Adam(model.parameters(), lr=0.01)

        model.train()
        print(f"Training {model_type}...")
        for epoch in range(50):  # Reduced epochs for speed
            optimizer.zero_grad()
            embeddings = model(x, edge_index_tensor, edge_weight_tensor if model_type == "GNN" else None)

            # Simple reconstruction loss
            reconstructed = torch.mm(embeddings, embeddings.t())
            target_similarity = torch.FloatTensor(similarity_matrix)
            loss = F.mse_loss(reconstructed, target_similarity)

            loss.backward()
            optimizer.step()

            if epoch % 10 == 0:
                print(f"Epoch {epoch}: Loss = {loss.item():.4f}")

        # Get final embeddings
        model.eval()
        with torch.no_grad():
            final_embeddings = model(x, edge_index_tensor, edge_weight_tensor if model_type == "GNN" else None)
            embeddings_np = final_embeddings.numpy()

        # Perform clustering on embeddings
        print("Performing clustering...")
        kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
        cluster_labels = kmeans.fit_predict(embeddings_np)

        # Evaluate clustering
        evaluation = evaluate_clustering(embeddings_np, cluster_labels, f"{feature_name}+{model_type}")

        return {
            'method': f"{feature_name} + {model_type}",
            'cluster_labels': cluster_labels,
            'embeddings': embeddings_np,
            'evaluation': evaluation,
            'features_shape': features_dense.shape,
            'n_edges': edge_index.shape[1],
            'similarity_stats': {
                'min': similarity_matrix.min(),
                'max': similarity_matrix.max(),
                'mean': similarity_matrix.mean()
            }
        }

    except Exception as e:
        print(f"Error in {feature_name} + {model_type} experiment: {str(e)}")
        return {
            'method': f"{feature_name} + {model_type}",
            'error': str(e),
            'features_shape': features.shape if hasattr(features, 'shape') else 'unknown'
        }
```

## Clustering Alternatif dengan DBSCAN

### Impor Pustaka DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.neighbors import NearestNeighbors
import matplotlib.pyplot as plt
import numpy as np
```

### Menemukan Parameter Epsilon (eps) Optimal untuk DBSCAN

```python
# Prepare the embeddings from your GNN experiment
embeddings = experiment1_results['embeddings']

# Step 1: Determine optimal epsilon parameter using k-distance graph
def find_optimal_epsilon(data, k=4):
    neigh = NearestNeighbors(n_neighbors=k)
    neigh.fit(data)
    distances, _ = neigh.kneighbors(data)
    k_distances = np.sort(distances[:, -1])

    plt.figure(figsize=(10, 6))
    plt.plot(k_distances)
    plt.title('K-Distance Graph (k=4)')
    plt.xlabel('Points sorted by distance')
    plt.ylabel(f'{k}-th nearest neighbor distance')
    plt.grid(True)
    plt.show()

    # The "elbow" point is a good candidate for epsilon
    return k_distances[int(len(k_distances)*0.95)]  # Using 95th percentile as default

optimal_eps = find_optimal_epsilon(embeddings)
print(f"\nSuggested epsilon value from k-distance graph: {optimal_eps:.4f}")
```

### Menjalankan Algoritma DBSCAN

```python
# Step 2: Run DBSCAN with selected parameters
dbscan = DBSCAN(eps=optimal_eps, min_samples=5)
dbscan_labels = dbscan.fit_predict(embeddings)
# Analyze DBSCAN results
n_clusters = len(set(dbscan_labels)) - (1 if -1 in dbscan_labels else 0)
n_noise = list(dbscan_labels).count(-1)

print("\nDBSCAN Results:")
print(f"Number of clusters: {n_clusters}")
print(f"Number of noise points: {n_noise}")
print(f"Percentage of noise: {n_noise/len(dbscan_labels)*100:.2f}%")
```

### Visualisasi Hasil Clustering DBSCAN

```python
# Step 3: Visualize the clusters and noise
# Reduce dimensionality for visualization using UMAP (better than PCA for visualization)
try:
    import umap
    reducer = umap.UMAP(random_state=42)
    vis_embeddings = reducer.fit_transform(embeddings)
    print("\nUsing UMAP for dimensionality reduction")
except ImportError:
    from sklearn.decomposition import PCA
    reducer = PCA(n_components=2)
    vis_embeddings = reducer.fit_transform(embeddings)
    print("\nUsing PCA for dimensionality reduction (install umap for better visualization)")

plt.figure(figsize=(12, 8))

# Plot clusters
unique_labels = set(dbscan_labels)
colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))]

for k, col in zip(unique_labels, colors):
    if k == -1:
        # Black used for noise
        col = [0, 0, 0, 1]
        label = 'Noise'
    else:
        label = f'Cluster {k}'

    class_member_mask = (dbscan_labels == k)

    xy = vis_embeddings[class_member_mask]
    plt.scatter(xy[:, 0], xy[:, 1], c=[col], s=30, alpha=0.6, label=label)

plt.title(f'DBSCAN Clustering (Estimated {n_clusters} clusters)\nNoise points: {n_noise} ({n_noise/len(dbscan_labels)*100:.1f}%)')
plt.legend()
plt.grid(True)
plt.show()
```
